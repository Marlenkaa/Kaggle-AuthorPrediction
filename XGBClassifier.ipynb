```python
import pandas as pd
from sklearn.model_selection import train_test_split
import xgboost
from sklearn import svm, tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
data = pd.read_csv('./INPUT/training_dataset.csv')
```

```python
data.head()
```

```python
# Nos quitamos el id
data = data.iloc[:,1:]
data.head()
```

```python
# Asignamos las features a la X
X = data.drop(columns='author')
y = data.author
```

```python
# Dividimos los datos en entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, shuffle=True)
```

```python
# Entrenamos el modelo

models = {
    'XGBC':xgboost.XGBClassifier(),
    'svm': svm.SVC(),
    'treeClassifier': tree.DecisionTreeClassifier(),
    'randomForest': RandomForestClassifier()
    }

for name,model in models.items():
    model.fit(X_train, y_train)
    y_pred= model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f'Accuracy of {model} is {acc}')
    cm = confusion_matrix(y_test, y_pred)
    print(f'Confusion Matrix of {model} is {cm}')
```

```python
# Entrenamos el modelo
model = xgboost.XGBClassifier()
data_trained = model.fit(X_train, y_train)
# Lo aplicamos para hacer la predicción
pred = data_trained.predict(X_test)
```
